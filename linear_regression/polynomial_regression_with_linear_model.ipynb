# Polynomial Regression with Linear Model

If our data is actually more complex than a simple straight line, a linear model will fit it badly, if we want our model to fit properly, we can add a feature in the dataset which will simply be our X values to the power of **n**.
Let's see how to do that.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(42)
```

We start by generating a dataset.

```python
n = 3
X = 6 * np.random.rand(200, 1) - 3
y = 0.5 * X**n + X/2 + np.random.randn(200, 1)

fig = plt.figure(figsize=(8,6))
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.show()
```

```python
X.shape
```

Now we need to add a feature to our data.  
New feature = $\large X_i^n$, **n** being the power value from our function.



```python
X_poly = np.append(X, X**n, 1)
print(X_poly[0:5])
```

We'll use the Normal Equation. Because it's short.

```python
def normal_equation(X, y):
    theta = np.linalg.inv(X.T @ X) @ (X.T @ y) # @ operator = Dot product
    return theta

theta = normal_equation(X_poly, y)
print(theta)
```

```python
predict_test = X_poly @ theta
```

```python
fig = plt.figure(figsize=(8,6))
plt.title('Dataset')
plt.scatter(X, y)
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.scatter(X, predict_test, color='red')
plt.show()
```
