# Batch Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(42) #Persistance in randomness.
```

We start by generating a dataset.

```python
X = 2.5 * np.random.rand(200, 1)
y = 5 + 3 * X + np.random.randn(200, 1)

fig = plt.figure(figsize=(8,6))
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.axis([0, 2.75, 0, 15])
plt.show()
```

To implement a gradient descent, we need to compute the gradient of the cost function with regards to each model parameters $\large\theta_j$, basically see how much the cost function changes when $\large\theta_j$ changes.
This is a partial derivative.
Here is the formula:

$\Large\frac {\partial } {\partial \theta _{j}}MSE\left( \theta \right) =\frac {2} {m}\sum _{i=1}^{m}\left( \theta ^{T}X^{i}-y^{i}\right) X_{j}^{i}$

However, in our case, we want to compute all these partial derivatives at once.
We will use the gradient vector noted $\large\nabla _{\theta }MSE\left( \theta \right)$ which contains all the partial derivative of the cost functions.  
  
$\Large\nabla _{\theta }MSE\left( \theta \right) =\left( \begin{matrix} \frac {\partial } {\partial \theta _{0}}MSE\left( \theta \right) \\ \frac {\partial } {0\theta _{1}}MSE\left( \theta \right) \\ \vdots \\ \frac {0} {\partial \theta _{n}}MSE\left( \theta \right) \end{matrix} \right) =\frac {2} {m}X^{T}\left( X\theta -y\right)$

```python
def batch_gradient_descent(X, y, learning_rate=0.1, n_iterations=1000, m=200):
    theta = np.random.randn(2,1)
    for iteration in range(n_iterations):
        gradients = 2/m * X.T @ (X @ theta - y)
        theta -= learning_rate * gradients
    return theta
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42) # Split the dataset
n_samples, _ = X_train.shape
n_samples_test, _ = X_test.shape
X_b_train = np.c_[np.ones((n_samples)), X_train]
X_b_test = np.c_[np.ones((n_samples_test)), X_test]
```

```python
theta = batch_gradient_descent(X_b_train, y_train)
print(theta)
```

```python
predict_test = X_b_test @ theta
```

```python
fig = plt.figure(figsize=(8,6))
plt.title(f'Intercept: {theta[0]} Slope: {theta[1]}')
plt.scatter(X_test, y_test)
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.scatter(X_test, predict_test, color='red')
plt.plot(X_test, predict_test, color='red')
plt.axis([0, 2.75, 0, 15])
plt.show()
```

# Magic!  
  
You could try modifying the learning_rate and number of iterations values.  
A learning_rate value set too low will take too long (relatively speaking) to converge to the minimum.  
A learning_rate value set too high will "overshoot" and might never find the minimum.
