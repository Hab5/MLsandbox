# Ridge Regression

One way to limit overfitting is the regularize our model (i.e to constrain it to a fewer degree of freedom it has)  
This technically isn't linear regression anymore, but I don't want to create another directory for a simple ridge regression, and it's similar enough.  
I duplicated the polynomial_regression_with_linear_model notebook, we will only change 1 line.  
Let's see how to do that.  

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(42)
```

We start by generating a dataset.

```python
poly_degree = 3
X = 6 * np.random.rand(200, 1) - 3
y = 0.5 * X**poly_degree + X/2 + 4 + np.random.randn(200, 1)

fig = plt.figure(figsize=(8,6))
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.show()
```

```python
X_poly = np.append(X, X**poly_degree, 1) # Polynomial features, see polynomial_regression_with_linear_model.ipynb
X_poly = np.c_[np.ones((200)), X_poly] # Adding bias
```

Ok, the only thing we have to do is add our regularization term to the normal equation (we could use any other linear model)
$\LARGE\theta =\left( X^{T}X \color{green} + \color{green} \lambda \color{green}I\right)^{-1}\cdot \left( X^{T}y\right)$

```python
def regularized_normal_equation(X, y, alpha):
    theta = np.linalg.inv(X.T @ X + alpha * np.identity(y.shape[1])) @ (X.T @ y) # @ operator = Dot product
    return theta

theta_low_alpha = regularized_normal_equation(X_poly, y, alpha=0.1)
theta_high_alpha = regularized_normal_equation(X_poly, y, alpha=100)
```

This take the form of an identity matrix multiplied by our parameter `alpha`.
The bigger alpha is, the less constrained is our model.

```python
predict_low_alpha = X_poly @ theta_low_alpha
predict_high_alpha = X_poly @ theta_high_alpha
```

```python
fig = plt.figure(figsize=(8,6))
plt.title('Dataset')
plt.scatter(X, y)
plt.xlabel("Independant Variable")
plt.ylabel("Dependant Variable")
plt.scatter(X, predict_low_alpha, color='red', label='Low Alpha')
plt.scatter(X, predict_high_alpha, color='orange', label='High Alpha')
plt.legend(loc='upper left')
plt.show()
```
